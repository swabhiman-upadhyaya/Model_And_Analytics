Q-1: Find the result of the given functions in:
    floor(), ciel, sqrt(), isqrt()

```python
import math

a = float(input("Enter an float number to find it's 'floor() & ciel():'"))
a_floor = math.floor(a)
print(f"The 'floor' of {a} is: {a_floor}")


a_ciel = math.ceil(a)
print(f"The 'ceil' of {a} is: {a_ciel}")


b = int(input("Enter an integer number to find it's 'sqrt()':"))
b_sqrt = math.sqrt(b)
print(f"The 'sqrt' of {b} is: {b_sqrt}")


c = int(input("Enter a number to check whether 'it is sqrt or not'"))
c_isqrt = math.isqrt(c)
print(f"The 'isqrt' of {c} is: {c_isqrt}")
```
