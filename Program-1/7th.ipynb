Q-7: Find the EigenValues and EigenVectors using Scipy *?

```python
import numpy as np
from scipy import linalg

arr = np.array ([
    [1, 2], 
    [3, 4]
])

eigenValues, eigenVectors = linalg.eig(arr)

print (f"Eigen Values = {eigenValues}\n")
print (f"Eigen Vectors = \n{eigenVectors}\n")

# ANOTHER METHOD: 
print ("BY ANOTHER METHOD: \n")

eigenValues2, eigenVectors2 = np.linalg.eig (arr)

print (f"Eigen Values2 = {eigenValues2}")
print (f"Eigen Values2 = \n{eigenVectors2}")
```
