Q-3: Usage of methods such as den(), eig() etc

```python
import numpy as np

arr = np.array ([
    [1, 2], 
    [3, 4]
])

det_arr = np.linalg.det(arr)
print (det_arr)
print()

eigenValues, eigenVectors = np.linalg.eig (arr)

print (f"EigenValues: {eigenValues}\n")

print (f"EigenValues: {eigenVectors}")
```
