##### Q-3: Demonstrate various data pre-processing techniques for a given dataset. WAP to compute

i: Reshaping the data

```python
import numpy as np

data = np.array([10, 20, 30, 40, 50, 60])

reshape_data = data.reshape(2, 3)
print(reshape_data)
```

ii: Filtering the data

```python
import pandas as pd

data = {
  "Name": ["Chikun", "Pikun", "Lipun"],
  "Age": [21, 22, 23],
  "Degree": ["MCA", "Hotel Mangement", "MCA"]
}

df = pd.DataFrame(data)
filtered_data = df[(df["Age"] >= 21) & (df["Degree"] == "MCA")]
print(filtered_data)
```

iii: Merging the data

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['A', 'B', 'C']
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Marks': [80, 90, 70]
})

merged_df = pd.merge(df1, df2, on="ID", how="outer")
print(merged_df)
```
