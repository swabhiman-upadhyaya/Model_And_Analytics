##### Q-2: WAP to compute Mean, Median, Mode, Variance, Standard deviation, using Dataset

```python
import pandas as pd
import numpy as np
from statistics import mode

df = pd.read_csv("world_famous_places_2024.csv")

data = df['Annual_Visitors_Millions']
# print(data)

print(f"Mean: {data.mean()}")
print(f"Median: {data.median()}")
print(f"Mode: {data.mode()[0]}")
print(f"Variance: {data.var()}")
print(f"Std deviation: {data.std()}")
```
