# Linear and Polynomial Regression Demonstration

This notebook demonstrates linear regression using the Boston Housing dataset and polynomial regression for vehicle fuel efficiency prediction.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Part 1: Linear Regression with California Housing Dataset
print("="*60)
print("LINEAR REGRESSION - CALIFORNIA HOUSING DATASET")
print("="*60)

# Load California Housing dataset (replacement for Boston Housing)
housing = fetch_california_housing()
X_housing = pd.DataFrame(housing.data, columns=housing.feature_names)
y_housing = housing.target

print("California Housing Dataset:")
print(f"Features: {list(housing.feature_names)}")
print(f"Target: {housing.target_names[0]} (Median house value in $100,000s)")
print(f"Dataset shape: {X_housing.shape}")

# Use MedInc (median income) as feature for simple linear regression
X_income = X_housing[['MedInc']]
X_train, X_test, y_train, y_test = train_test_split(X_income, y_housing, test_size=0.2, random_state=42)

# Train linear regression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)

# Predict
y_pred = lin_reg.predict(X_test)

# Evaluate
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"\nLinear Regression Results:")
print(f"Coefficient: {lin_reg.coef_[0]:.2f}")
print(f"Intercept: {lin_reg.intercept_:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R² Score: {r2:.2f}")

print("\nSample Predictions:")
for i in range(5):
    print(f"Actual: {y_test[i]:.2f}, Predicted: {y_pred[i]:.2f}")

# Part 2: Polynomial Regression for Vehicle Fuel Efficiency
print("\n" + "="*60)
print("POLYNOMIAL REGRESSION - VEHICLE FUEL EFFICIENCY")
print("="*60)

# Create synthetic dataset for vehicle fuel efficiency
np.random.seed(42)
n_samples = 100
speed = np.random.uniform(20, 120, n_samples)  # Speed in km/h
# Fuel efficiency (km/L) with polynomial relationship
fuel_efficiency = 15 - 0.05 * speed + 0.0005 * speed**2 + np.random.normal(0, 1, n_samples)

# Create DataFrame
df_fuel = pd.DataFrame({'speed': speed, 'fuel_efficiency': fuel_efficiency})

print("Vehicle Fuel Efficiency Dataset (Synthetic):")
print(f"Features: Speed (km/h)")
print(f"Target: Fuel Efficiency (km/L)")
print(f"Dataset shape: {df_fuel.shape}")
print("\nFirst 5 samples:")
print(df_fuel.head())

# Split data
X_fuel = df_fuel[['speed']]
y_fuel = df_fuel['fuel_efficiency']
X_train_f, X_test_f, y_train_f, y_test_f = train_test_split(X_fuel, y_fuel, test_size=0.2, random_state=42)

# Linear Regression for comparison
lin_reg_f = LinearRegression()
lin_reg_f.fit(X_train_f, y_train_f)
y_pred_lin = lin_reg_f.predict(X_test_f)

# Polynomial Regression (degree 2)
poly_features = PolynomialFeatures(degree=2)
X_train_poly = poly_features.fit_transform(X_train_f)
X_test_poly = poly_features.transform(X_test_f)

poly_reg = LinearRegression()
poly_reg.fit(X_train_poly, y_train_f)
y_pred_poly = poly_reg.predict(X_test_poly)

# Evaluate both
mse_lin = mean_squared_error(y_test_f, y_pred_lin)
r2_lin = r2_score(y_test_f, y_pred_lin)

mse_poly = mean_squared_error(y_test_f, y_pred_poly)
r2_poly = r2_score(y_test_f, y_pred_poly)

print(f"\nLinear Regression Results:")
print(f"MSE: {mse_lin:.2f}, R²: {r2_lin:.2f}")

print(f"\nPolynomial Regression (Degree 2) Results:")
print(f"MSE: {mse_poly:.2f}, R²: {r2_poly:.2f}")

print("\nSample Predictions (Linear vs Polynomial):")
for i in range(5):
    speed_val = X_test_f.iloc[i, 0]
    actual = y_test_f.iloc[i]
    pred_lin = y_pred_lin[i]
    pred_poly = y_pred_poly[i]
    print(f"Speed: {speed_val:.1f} km/h, Actual: {actual:.2f}, Linear: {pred_lin:.2f}, Poly: {pred_poly:.2f}")

print("\n" + "="*60)
print("SUMMARY")
print("="*60)
print("Linear Regression (California Housing):")
print(f"  Predicts house prices based on median income")
print(f"  R² Score: {r2:.2f}")

print("\nPolynomial Regression (Fuel Efficiency):")
print(f"  Models non-linear relationship between speed and fuel efficiency")
print(f"  Linear R²: {r2_lin:.2f}, Polynomial R²: {r2_poly:.2f}")
if r2_poly > r2_lin:
    print("  Polynomial regression performs better for this non-linear relationship")
else:
    print("  Results may vary due to synthetic data")
```
