##### Q-7: Usage of different methods in Matplotlib

plt.plot()	Line graph
plt.scatter()	Scatter plot
plt.bar()	Bar chart
plt.barh()	Horizontal bar chart
plt.hist()	Histogram
plt.pie()	Pie chart
plt.xlabel()	X-axis label
plt.ylabel()	Y-axis label
plt.title()	Title of graph
plt.legend()	Show legend
plt.grid()	Show grid
plt.show()	Display the plot

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 30, 40]

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Graph")
plt.show()
```
