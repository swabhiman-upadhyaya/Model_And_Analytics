##### Q-3: Compare an array using Numpy with a series using pandas

```python
import numpy as np
import pandas as pd

ser1 = pd.Series([10, 20, 30], index=['a', 'b', 'c'])
ser2 = pd.Series([1, 2, 3], index=['b', 'c', 'd'])
print(f"{ser1 + ser2}\n")


arr1 = np.array([10, 20, 30])
arr2 = np.array([1, 2, 3])
print(f"{arr1 + arr2}")
```
