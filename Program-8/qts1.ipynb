Q-1: To classify the data points using the k-nearest neighboiurs algorithm and study the effect of different values of k.
theory: 
k-nearest neighbours (knn) is a supervised machine learning algorithm used for classification and regression tasks. it is a non-parametric and lazy learning algorithm, meaning it does not build an explicit model during the training phase. instead, it stores the entire training dataset and perform computation only at  prediction time.

```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

from sklearn.model_selection import train_test_split

from sklearn.neighbors import  KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
iris=load_iris()
x=iris.data
y=iris.targe
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.2,random_state=42
)
```

```python
# create KNN classifier
k=5
knn=KNeighborsClassifier(n_neighbors=k)

```

```python
# train the model
knn.fit(x_train,y_train)
```

```python
# make predictions
y_pred=knn.predict(x_test)
```

```python
# evalute the model 
accuracy=accuracy_score(y_test,y_pred)
print("Accuracy: ",accuracy)
```

```python
print("\nClassification Report :")
print(classification_report(y_test,y_pred,
target_names=iris.target_names))
```

```python
# Accuracy VS Value Experiment

accuracy_list=[]

for k in range(1,21):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred = knn.predict(x_test)
    accuracy_list.append(accuracy_score(y_test,y_pred))
plt.plot(range(1,21),accuracy_list,marker='o')
plt.xlabel("value of K")
plt.ylabel("Accuracy")

```
