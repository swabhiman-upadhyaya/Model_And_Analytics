## implement simple and multiple linear Regression models.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

blobs=make_blobs(n_samples=100, centers=3, n_features=2, random_state=42) #creating custom dataset
X=blobs[0] #taking only the features
kmeans1=KMeans(n_clusters=1, random_state=42)
kmeans1.fit(X)
print("Cluster centers:\n", kmeans1.cluster_centers_)
print("Labels:\n", kmeans1.labels_)

plt.scatter(X[:, 0], X[:, 1], c=kmeans1.labels_, cmap='viridis')
plt.scatter(kmeans1.cluster_centers_[:, 0], kmeans1.cluster_centers_[:, 1], c='red', marker='^', s=200)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
kmeans2=KMeans(n_clusters=3, random_state=42)
kmeans2.fit(X)
print("Cluster centers:\n", kmeans2.cluster_centers_)
print("Labels:\n", kmeans2.labels_)

plt.scatter(X[:, 0], X[:, 1], c=kmeans2.labels_, cmap='viridis')
plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1], c='red', marker='^', s=200)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

```python
kmeans3=KMeans(n_clusters=5, random_state=42)
kmeans3.fit(X)
print("Cluster centers:\n", kmeans3.cluster_centers_)
print("Labels:\n", kmeans3.labels_)

plt.scatter(X[:, 0], X[:, 1], c=kmeans3.labels_, cmap='viridis')
plt.scatter(kmeans3.cluster_centers_[:, 0], kmeans3.cluster_centers_[:, 1], c='red', marker='^', s=200)
plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
