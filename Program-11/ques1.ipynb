### Implement Support Vector Machine for a dataset and compare the accuracy by applying the following kernel functions i.Linear ii.Polynomial iii.RBF 

```python
from sklearn.datasets import load_iris
from sklearn import svm
import matplotlib.pyplot as plt
import numpy as np

iris=load_iris()
X=iris.data[:,:2]
y=iris.target

```

```python
x_min,x_max=X[:,0].min()-1,X[:,0].max()+1
y_min,y_max=X[:,1].min()-1,X[:,1].max()+1
h= (x_max / x_min)/100
xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
X_plot=np.c_[xx.ravel(),yy.ravel()]
```

```python
C=1.0
svc_classifier=svm.SVC(kernel='linear',C=C).fit(X,y)
z=svc_classifier.predict(X_plot)
z=z.reshape(xx.shape)
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) 
plt.contourf(xx,yy,z,cmap=plt.cm.coolwarm,alpha=0.8)
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm,edgecolors='k')
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Support vector classifier with linear kernel")
plt.xlim(xx.min(),xx.max())
```

```python
svc_classifier=svm.SVC(kernel='rbf',gamma='auto',C=C).fit(X,y)
z=svc_classifier.predict(X_plot)
z=z.reshape(xx.shape)
plt.figure(figsize=(15,5))
plt.subplot(121) 
plt.contourf(xx,yy,z,cmap=plt.cm.coolwarm,alpha=0.8)
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm,edgecolors='k')
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Support vector classifier with rbf kernel")
plt.xlim(xx.min(),xx.max())
```

```python
svc_classifier=svm.SVC(kernel='poly',C=C).fit(X,y)
z=svc_classifier.predict(X_plot)
z=z.reshape(xx.shape)
plt.figure(figsize=(15,5))
plt.subplot(1,2,1) 
plt.contourf(xx,yy,z,cmap=plt.cm.coolwarm,alpha=0.8)
plt.scatter(X[:,0],X[:,1],c=y,cmap=plt.cm.coolwarm,edgecolors='k')
plt.xlabel("sepal length")
plt.ylabel("sepal width")
plt.title("Support vector classifier with poly kernel")
plt.xlim(xx.min(),xx.max())
```

```python
#accuracy
from sklearn.metrics import accuracy_score
y_pred=svc_classifier.predict(X)
accuracy=accuracy_score(y,y_pred)
print("Accuracy:",accuracy)
```
